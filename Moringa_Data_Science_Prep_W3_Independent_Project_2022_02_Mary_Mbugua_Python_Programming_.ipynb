# Importing Libraries

```python
# Importing the libraries for the task
import pandas as pd
import numpy as np
import datetime
import time
```

# Loading the Datasets

```python
# Loading the Telcom dataset 1 that we are going to use and displaying each dataframe created
dataframe1 = pd.read_csv('Telcom_dataset.csv')
dataframe1.head()
```

```python
# Loading the Telcom dataset 2 and displaying the dataframe created
dataframe2 = pd.read_csv('Telcom_dataset2.csv')
dataframe2.head()
```

```python
# Loading the Telcom dataset 3 and displaying the dataframe created
dataframe3 = pd.read_csv('Telcom_dataset3.csv')
dataframe3.head()
```

```python
# Loading the cells_geo dataset and displaying the dataframe created
cells_geo = pd.read_csv('cells_geo.csv')
cells_geo.head()
```

# Data Cleaning

```python
# Renaming columns in dataframe1
dataframe1.rename(columns = {'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME'},inplace=True)
dataframe1.head()
```

```python
# Renaming columns in dataframe3
dataframe3.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT','CELLID':'CELL_ID', 'SIET_ID':'SITE_ID'},inplace=True)
dataframe3.head()
```

```python
# Renaming columns in dataframe2
dataframe2.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'},inplace=True)
dataframe2.head()
```

```python
# The columns in the cells geo dataframe need to be split so each of them has its own column.
# Spliting the columns using ";" as the delimiter and renaming them as with the data provided

cells_geo = cells_geo[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].apply(lambda x:pd.Series([i for i in reversed(x.lower().split(';'))]))
cells_geo.rename(columns = {0:'SITE_ID',1:"CELL_ID",2:"AREA",3:"REGION",4:"LATITUDE",5:"LONGITUDE",6:"ZONENAME",7:"DECOUPZONE",8:"LOCALIZATION",9:"STATUS",10:"CITY"},inplace=True)
cells_geo.head(30)
```

```python
# Concatnating the Telcom dataframes into one dataframe
dataframes = [dataframe1,dataframe2,dataframe3]
dataframes = pd.concat(dataframes)
dataframes
```

```python
# confirming whether business is done during business hours or on home hours
# this is done by convert the 'DATE_TIME' column type to datetime then change  the times such that 8:00-17:00 is working hours
dataframes['DATE_TIME'] = pd.to_datetime(dataframes['DATE_TIME'], format="%Y-%m-%d %H:%M:%S")
dataframes.head()
```

```python
# spliting date and time on different columns so as to filter out times
dataframes['Dates'] = pd.to_datetime(dataframes['DATE_TIME'], format='%D:%M:%Y').dt.date
dataframes['Hours'] = pd.to_datetime(dataframes['DATE_TIME'], format='%D:%M:%Y').dt.time
dataframes
```

```python
# converting the 'Hours' column back to datetime format so as to filter out the working hours
dataframes['Hours'] = pd.to_datetime(dataframes['Hours'], format='%H:%M:%S' ).apply(pd.Timestamp)
dataframes
```

```python
# summary of statistics by the Product value
dataframes["VALUE"].groupby(dataframes["PRODUCT"]).describe()
```

```python
# summary of statistics in cells geo by City
cells_geo['CELL_ID'].groupby(cells_geo["AREA"]).describe()
```

```python
# filtering out the times by using 08:00-17:00 for business hours and the opposite for home hours
working = (dataframes['Hours'] > '1900-01-01 08:00:00') & (dataframes['Hours'] <= '1900-01-01 17:00:00')
home = (dataframes['Hours'] > '1900-01-01 17:00:00') & (dataframes['Hours'] < '1900-01-01 08:00:00')
print(dataframes.loc[home])
```
